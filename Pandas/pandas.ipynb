```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("khcam.csv")
data = data.groupby(by = "分局").size()
plt.rcParams["font.family"] = "DFKai-SB"

data.plot(kind = "bar" , figsize = (10,6),fontsize =18)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("khcam.csv")
data = data.groupby(by = "分局").size().sort_values(ascending = False)
plt.rcParams["font.family"] = "DFKai-SB"

data.plot(kind = "bar" , figsize = (10,6),fontsize =18)
```
